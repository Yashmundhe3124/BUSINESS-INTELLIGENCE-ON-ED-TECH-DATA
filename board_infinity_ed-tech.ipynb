#### create 4 columns with created date 
#### day month year year month

```python
df['created'] = pd.to_datetime(df['created'])
df['date_created'] = df['created'].dt.date
df['Day'] = df['created'].dt.day_name()
df['Month'] = df['created'].dt.month_name()
df['MonthYear'] = df['created'].dt.strftime('%B-%Y')
df.head()
```

```python
df.drop(['created'],axis=1,inplace=True)
```

```python
q1=516.000000
q3=4913.500000
iqr=q3-q1
iqr
lower_limit=q1 - 1.5 * iqr
upper_limit=q3 + 1.5 * iqr
df1=df[(df.num_subscribers<upper_limit) & (df.num_subscribers>lower_limit)]
```
